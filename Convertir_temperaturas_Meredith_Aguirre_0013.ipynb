### 1. Fórmulas de Conversión

Para realizar estas operaciones, utilizamos las siguientes relaciones matemáticas:

* **De Grados Celsius (°C) a Grados Fahrenheit (°F):**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(O también se puede usar $1.8$ en lugar de $9/5$)*.

* **De Grados Fahrenheit (°F) a Grados Celsius (°C):**
    $$°C = \frac{(°F - 32) \times 5}{9}$$
    *(O también se puede expresar como $(°F - 32) / 1.8$)*.

---

### 2. Ejemplos Resueltos Paso a Paso

#### Ejemplo 1: Convertir 25°C a °F (Temperatura ambiente agradable)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$.
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar 32:** $45 + 32 = 77$.
* **Resultado:** 25°C equivalen a **77°F**.

#### Ejemplo 2: Convertir 98.6°F a °C (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$.
2.  **Sustituir el valor:** $°C = (98.6 - 32) / 1.8$.
3.  **Restar 32 primero (paréntesis):** $98.6 - 32 = 66.6$.
4.  **Dividir entre 1.8:** $66.6 / 1.8 = 37$.
* **Resultado:** 98.6°F equivalen a **37°C**.

### 3. Código en Python

Aquí tienes el programa interactivo siguiendo todos tus requerimientos técnicos:

```python
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius a convertir: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n=== CONVERSOR DE TEMPERATURA ===")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("¡Oscar Flores NL = 19!")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Llamada inicial al menú
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
* **`while True`**: Crea un bucle infinito que permite que el programa se repita hasta que el usuario decida salir.
* **`input()`**: Captura la respuesta del usuario.
* **`float()`**: Convierte el texto ingresado en un número decimal para poder realizar cálculos.
* **`if/elif/else`**: Estructura de control que decide qué función ejecutar según el número presionado.
* **`break`**: Rompe el bucle `while` y finaliza la ejecución cuando se elige la opción 3.
* **`f-strings` (ej. `f"{f:.2f}"`)**: Se utilizan para mostrar los resultados con solo 2 decimales para mayor limpieza visual.

````python
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius a convertir: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n=== CONVERSOR DE TEMPERATURA ===")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("¡Meredith Aguirre NL = 2!")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Llamada inicial al menú
if __name__ == "__main__":
    menu()
````
